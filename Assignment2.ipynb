## Part B: Questions to Sune's talk on explanatory data visualization
### Three key elements to keep in mind when you design an explanatory visualization?

The three key elements are listed below, and then further detailes are described
- Start with a question 
- Allow exploration 
- Know your readers 

**Start with a question**

For this part one has to keep in mind questions that can help guide the project. 
What is it we, as creator of the visulzation want to commucinate for the reader, where are we going with this. This is what reader/user have to understand from this project, and their key takeaways 

**Allow exploration**

As users learn more from interactions and being engaged, the visualzations should reflect the same principles. It should allow the user to interact, and create their own insights from the visualzation, thus keeping them more engaged. In order to do so, a visualzation should follow the listed approach: 
- Provide the user with an overview - high level view of data (1)
- Then allow the user to zoom and filter, and choose certain elements and only show those (2)
- Details on demand, which allow the user to explore all the data (3)


**Know your readers**
 
Below are the listed quenstions and thoughts the creator of a visualzation should have in mind:
- What is the technical level and capabilities of intented audiance audiance? 
- How much prior knowlegde do they have about the dataset? As the creator one should always know
- How easy is to understand the visualzation? Don't choose to comprehensive visuals, that a user, won't fully understand
- What is goal or the intented audiance. What should their key takeaways be? 
- How much time is the inteted user expected to use on your visualzation, before creating the desired overview and forming insights? A quick glance, a short seesion, a interaction or more time?  

It is very crucial to understand who the inteded readers are. Are they techical experts or set of friends. Each group have different levels of expertise allowing them do understand and to goo deeper in depth, which should be incorporated in the visualzation. [3]

### Part 2A: Random forest binary classification.

In this section a random forrest classifier is build to distinguish between crime category "Burglary" and "Vehicle theft" using spatio-temporal features in the first model. In the second model weather data are added to see if this improves the accuracy of the classifier. 

The data preperation for both models is done first such that the same crimes are picked for training and test set in both models to make the two classifiers as compareable as possible in relation to the effect of the added weather data.
The data sets are first merged using the date and time from each dataset. 

Secondly the rows in which the wind direction is missing are dropped. In this case this only leads to dropping one of the incidents for the two investigated categories and therefore is not expected to impact the final classifier.    

```python
#data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv") 
#weather_data = pd.read_csv("weather_data.csv")
weather_data['date'] = pd.to_datetime(weather_data.date).dt.tz_localize(None)
data['datetime']=(pd.to_datetime(data['Date'] + ' ' + data['Time'])).round('60min')
merge=pd.merge(data,weather_data, how='inner', left_on='datetime', right_on='date')
merge['month'] = pd.DatetimeIndex(merge['datetime']).month
merge['hour'] = pd.DatetimeIndex(merge['datetime']).hour
merge = merge[merge['wind_direction'].notna()]
```

The data is hereafter balanced to avoid bias in the classifier. To use unbalanced training data could potentially lead to the classifier classify all data as the majority of the data e.g. if 80% of the crimes were Buglary and the classifier would classify all as Burglary, the accuracy would be 80%. However even though the accuracy would be fairly high this would not be a very useful model for determining the type of crime. 

To balance the data, the same amount of crimes are taken from both categories. As it is intented to include as much data as possible, all data from the category having the fewest incidents are included, and afterwards a random sample of equal size are picked from the category having the most data. The downside of this method is that some data is not included in the model. Other methods for handling unbalanced data could also have been chosen to avoid not including data. 

```python
pick = np.minimum(sum(merge.Category == 'BURGLARY'),sum(merge.Category == 'VEHICLE THEFT'))
Burglary = pd.DataFrame()
Vehicle = pd.DataFrame()
Burglary = merge.loc[merge['Category'] == 'BURGLARY'][['PdDistrict','DayOfWeek','month','hour','temperature','humidity','wind_speed','wind_direction','pressure','weather','Category']].sample(n=pick, random_state=42,replace=False)
Vehicle = merge.loc[merge['Category'] == 'VEHICLE THEFT'][['PdDistrict','DayOfWeek','month','hour','temperature','humidity','wind_speed','wind_direction','pressure','weather','Category']].sample(n=pick, random_state=42,replace=False)
df1 = pd.concat([Burglary,Vehicle],axis=0,ignore_index=False)
df1 = df1.reset_index(drop=True)
```

```python
WO_weather = pd.concat([pd.get_dummies(df1['DayOfWeek']),pd.get_dummies(df1['PdDistrict']),df1['month'],df1['hour'],df1['Category']],axis=1)
W_weather = pd.concat([pd.get_dummies(df1['DayOfWeek']),pd.get_dummies(df1['PdDistrict']),df1.iloc[:,2:9],pd.get_dummies(df1['weather']),df1['Category']],axis=1)
```

The two data sets, without weather data (WO_weather) and with weather data (W_weather), is not preparred for training and testing. Both data sets utilize one-hot encoding of the features: "DayOfWeek", "PdDistric". Additionally the weather dicribtion in the weather dataset is also one-hot encoded. 
The one-hot encoding makes the fitting of the models less efficient as a feature for each day of the week, each Pd district and each weather describtion are added to the data. However since there is no meaningful ordinality in the features it is chosen to done by one-hot encoding.

The data is first splitted into training and test set, where the size of the test set is chosen to be 10 %. Secondly cross validation is used to train the model in order to try to avoid overfitting and a grid search are added to tune the random forrest classication parameters.  

```python
train_features_WO, test_features_WO, train_labels_WO, test_labels_WO = train_test_split(WO_weather.iloc[:,:-1], WO_weather.iloc[:,-1], test_size = 0.1, random_state = 42)
train_features_W, test_features_W, train_labels_W, test_labels_W = train_test_split(W_weather.iloc[:,:-1], W_weather.iloc[:,-1], test_size = 0.1, random_state = 42)
```

The train and testsets are now used and their accuraucy in prediction are determined

```python
rfc_WO = RandomForestClassifier(n_jobs=-1,random_state=42,oob_score = True) 
param_grid = {
    'n_estimators': [100,500,1000],
    'max_features':['auto', 'sqrt']}

CV_rfc_WO = GridSearchCV(rfc_WO,param_grid=param_grid,scoring ='accuracy', cv= 10)
CV_rfc_WO.fit(train_features_WO, train_labels_WO)
print('\n',CV_rfc_WO.best_estimator_)
```

```python
print("accuracy training set =", np.mean(CV_rfc_WO.predict(train_features_WO) == train_labels_WO))
print("accuracy test set =", np.mean(CV_rfc_WO.predict(test_features_WO) == test_labels_WO))
```

The model without the weather data performs with an accuracy of 57.2% in the test set. As there is an equal amount of incidents a baseline model would have an accuracy of 50%. So overall the model is not very good at determining the category of the crime only based on time and place. 


### Part 2B: Info from weather features.

The model is fitted once again with the weather data included:

```python
rfc_W = RandomForestClassifier(n_jobs=-1,random_state=42,oob_score = True) 
param_grid = {
    'n_estimators': [100,500,1000],
    'max_features':['auto', 'sqrt']}

CV_rfc_W = GridSearchCV(rfc_W,param_grid=param_grid,scoring ='accuracy', cv= 10)
CV_rfc_W.fit(train_features_W, train_labels_W)
print('\n',CV_rfc_W.best_estimator_)
```

```python
print("accuracy training set =", np.mean(CV_rfc_W.predict(train_features_W) == train_labels_W))
print("accuracy test set =", np.mean(CV_rfc_W.predict(test_features_W) == test_labels_W))
```

The accuracy now improved to 60.8% which imposes that the weather does have an impact on which types of crimes that are commited. 
In both models the accuracy in the training sets are significantly higher than in the test set. That the training set has a higher accuracy is expected but that the diffence is so high could be a sign of overfitting.
Additionnally the grid search could also have been done with more parameter combinations, but as the time consumption for tuning the randomforest is high it was not carried out. The process could however have been speeded up by utilizing label encoding instead of one-hot encoding.
